```python
import instaloader
```

```python
from instaloader import Instaloader, Profile
```

```python
L = Instaloader()
```

```python
username = 'lsudominance'
```

```python
profile = Profile.from_username(L.context, username)
posts = profile.get_posts()
```

```python
owner_username = []
owner_id = []
date = []
url = []
typename = []
likes = []
comments = []


for post in posts:
    owner_username.append(post.owner_username)
    owner_id.append(post.owner_id)
    date.append(post.date)
    url.append(post.url)
    typename.append(post.typename)
    likes.append(post.likes)
    comments.append(post.comments)
```

```python
import pandas as pd

test_df1 = pd.DataFrame({"owner_username":owner_username,
                        "owner_id":owner_id,
                        "date":date,
                        "url":url,
                        "typename":typename,
                        "likes":likes,
                        "comments":comments
                       })
```

```python
test_df1
```

```python
test_df1.to_csv("lsudominance.csv")
```
